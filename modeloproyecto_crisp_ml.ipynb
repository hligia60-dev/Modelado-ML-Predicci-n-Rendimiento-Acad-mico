```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset oficial
df = pd.read_csv('student_dataset_10000_rows.csv')
print("--- PRIMERAS FILAS ---")
print(df.head())
print("\n--- INFORMACIÓN GENERAL ---")
print(df.info())
```

## 🛠️ Fase 3: Data Preparation (Preparación de los Datos)

En esta fase adecuamos la estructura de los datos para cumplir con las necesidades de los algoritmos de aprendizaje supervisado.

### 📐 Regla de Negocio (Criterio Oficial MINERD)
Para dar cumplimiento a los sistemas de evaluación del aprendizaje, se crea la variable objetivo binaria **`Riesgo_Academico`** mediante ingeniería de características:
* **Clase 1 (En Riesgo / Reprobado):** Estudiantes cuya calificación final es menor a 70 puntos ($exam\_score < 70$).
* **Clase 0 (Regular / Promovido):** Estudiantes con calificaciones iguales o superiores a 70 puntos ($exam\_score \ge 70$).

A continuación, se aíslan las variables predictoras ($X$) y se segmentan los dos vectores objetivos ($y\_lineal$ e $y\_logistica$) para alimentar la fase de modelado concurrente de manera limpia.


```python
import pandas as pd
import numpy as np 

# Crear la variable dependiente binaria de Riesgo Académico usando numpy de forma segura
df['Riesgo_Academico'] = np.where(df['exam_score'] < 70, 1, 0)

# Definir variables predictoras (X) eliminando las variables objetivo del set de características
X = df[['study_hours', 'attendance', 'sleep_hours', 'internet_usage', 'assignments_completed', 'previous_score']]

# Definir targets (Y)
y_lineal = df['exam_score']
y_logistica = df['Riesgo_Academico']

print("📊 Distribución de estudiantes en Riesgo:")
print(df['Riesgo_Academico'].value_counts())
print("\n¡Preparación de datos finalizada con éxito!")
```

## 🚀 Fase 4 & 5: Modeling & Evaluation (Modelado y Evaluación)

Procedemos a realizar la partición de los datos en conjuntos de **Entrenamiento (80%)** y **Prueba (20%)** utilizando una semilla de aleatoriedad garantizada (`random_state=42`) para la reproducibilidad del experimento.

### 🏗️ Modelos a Desplegar de Forma Concurrente:
1. **Regresión Lineal Múltiple:** Utilizada para modelar la tendencia continua de las calificaciones. Se evaluará mediante el Coeficiente de Determinación ($R^2$).
2. **Regresión Logística Binaria:** Utilizada para calcular la probabilidad de riesgo. Se evaluará mediante la métrica de exactitud (*Accuracy*).

### 💾 Serialización (Exportación del Conocimiento)
Una vez entrenados, exportamos los modelos a archivos estructurados `.pkl` usando `joblib` para que nuestro Dashboard interactivo de Streamlit los consuma en tiempo real.


```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import r2_score, accuracy_score
import joblib

# --- 1. Regresión Lineal (Calificación) ---
X_train_lin, X_test_lin, y_train_lin, y_test_lin = train_test_split(X, y_lineal, test_size=0.2, random_state=42)
modelo_lin = LinearRegression()
modelo_lin.fit(X_train_lin, y_train_lin)

pred_lin = modelo_lin.predict(X_test_lin)
print("--- Métrica Regresión Lineal ---")
print(f"R2 Score: {r2_score(y_test_lin, pred_lin):.4f}")

# --- 2. Regresión Logística (Riesgo Académico) ---
X_train_log, X_test_log, y_train_log, y_test_log = train_test_split(X, y_logistica, test_size=0.2, random_state=42)
modelo_log = LogisticRegression(max_iter=1000)
modelo_log.fit(X_train_log, y_train_log)

pred_log = modelo_log.predict(X_test_log)
print("\n--- Métrica Regresión Logística ---")
print(f"Accuracy: {accuracy_score(y_test_log, pred_log):.4f}")

# --- 3. Exportar Modelos ---
joblib.dump(modelo_lin, 'modelo_lineal_ufhec.pkl')
joblib.dump(modelo_log, 'modelo_logistica_ufhec.pkl')
print("\n¡Modelos guardados con éxito en la carpeta raíz del proyecto!")
```
